```python
import random

# Generate a list of 100 random numbers between 1000 and 10000
data = [random.randint(1000, 10000) for _ in range(1000)]
# data
```

```python
import numpy as np

mean_value = np.mean(data)
print("Mean:", mean_value)

median_value = np.median(data)
print("Median:", median_value)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random numbers
data = np.random.randn(1000)  # Generate 1000 random numbers from a normal distribution

# Calculate the standard deviation
std_dev = np.std(data)

# Plot the data
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black')

# Add a line for the mean
plt.axvline(np.mean(data), color='red', linestyle='dashed', linewidth=2, label='Mean')

# Add lines for the standard deviation
plt.axvline(np.mean(data) + std_dev, color='green', linestyle='dashed', linewidth=2, label='+1 Std Dev')
plt.axvline(np.mean(data) - std_dev, color='green', linestyle='dashed', linewidth=2, label='-1 Std Dev')

# Add labels and title
plt.title('Histogram of Random Numbers with Standard Deviation')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```
